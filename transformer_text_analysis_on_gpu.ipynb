```
## https://juliapackages.com/p/transformers
using Transformers.TextEncoders
using Transformers.HuggingFace
```

```
## load the hugging face model bert-base-uncased
## https://huggingface.co/google-bert/bert-base-uncased
textencoder, bert_model = hgf"bert-base-uncased"
```

```
## run nvidia-smi again
command = `nvidia-smi`
run(command)
```

```
## Define some strings of text
text1 = "Peter Piper picked a peck of pickled peppers"
text2 = "Fuzzy Wuzzy was a bear"
```

```
# Form a batch of contiguous sentences
text = [[ text1, text2 ]]
# Tokenize + pre-process (add special tokens + truncate / padding + one-hot encode)
sample = encode(textencoder, text)
```

```
## visualize the text in the way that it is being passed into the model
reshape(decode(textencoder, sample.token), :)
```

```
## check that text has been encoded in a particular way that can be parsed by the model
@assert reshape(decode(textencoder, sample.token), :) == [
    " [CLS]", " peter", " piper", " picked", " a", " peck", " of", " pick", "led", " peppers", " [SEP]",
    " fuzzy", " wu", "zzy",  " was", " a", " bear", " [SEP]"
]
```

```
## Pass the sample through the model and look at the output
## Why is it shaped this way??
bert_features = bert_model(sample).hidden_state
```

```
using LinearAlgebra

# Extract embeddings for tokens "peck" and "pick"
println(size(bert_model(sample).hidden_state))

token_features = bert_model(sample).hidden_state[:, :, 1]
peck_vector = token_features[:, 6]  # Assuming "peck" is the 6th token
pick_vector = token_features[:, 8]  # Assuming "pick" is the 8th token

# Compute similarity
similarity = dot(peck_vector, pick_vector) / (norm(peck_vector) * norm(pick_vector))
println("Similarity between 'peck' and 'pick': ", similarity)
```

```
text3 = "Peter Piper picked a peck of pickled peppers"
text4 = "He picked at the food"
text5 = "Pick any item you like"
text6 = "I cannot pick a favorite"

# Encode text samples
encode3 = encode(textencoder, [[text3]])
encode4 = encode(textencoder, [[text4]])
encode5 = encode(textencoder, [[text5]])
encode6 = encode(textencoder, [[text6]])

# Print how the samples are encoded
println(reshape(decode(textencoder, encode3.token), :))
println(reshape(decode(textencoder, encode4.token), :))
println(reshape(decode(textencoder, encode5.token), :))
println(reshape(decode(textencoder, encode6.token), :))

# Extract features for each sentence and compare
features3 = bert_model(encode3).hidden_state
features4 = bert_model(encode4).hidden_state
features5 = bert_model(encode5).hidden_state
features6 = bert_model(encode6).hidden_state

println(size(features3), size(features4), size(features5), size(features6))


```

Examine the similarity between the context/usage of the same words in the different sentences above.

```
# Analyze similarity between tokens

text3_pick_vector1 = features3[:, 4, 1] # Assuming "picked" is the 4th toekn
text3_pick_vector2 = features3[:, 8, 1] # Assuming "pick" is the 8th token
text4_pick_vector = features4[:, 3, 1]  # Assuming "picked" is the 3rd token
text5_pick_vector = features5[:, 2, 1]  # Assuming "pick" is the 2nd token
text6_pick_vector = features6[:, 4, 1]  # Assuming "pick" is the 4th token

typeof(text5_pick_vector)

# Compute similarity

sim_34 = dot(text3_pick_vector1, text4_pick_vector) / (norm(text3_pick_vector1) * norm(text4_pick_vector))
sim_56 = dot(text5_pick_vector, text6_pick_vector) / (norm(text5_pick_vector) * norm(text6_pick_vector))

println("Similarity between 'picked' and 'picked' in text3 and text4: ", sim_34)
println("Similarity between 'pick' and 'pick' in text5 and text6: ", sim_56)
```

```
# Define a function that computes the similarity between two vectors

function compute_similarity(vec1, vec2)
    ans = dot(vec1, vec2) / (norm(vec1) * norm(vec2))
    return ans
end

function compute_all_similarities(list)
    n = length(list)
    for i in 1:n
        for j in i+1:n
            ans = compute_similarity(list[i], list[j])
            println("Similarity between vector $i and $j is $ans")
        end
    end
end

```
